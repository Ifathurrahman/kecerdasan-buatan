````python
import json

def soal_1():
    reverse_sorted = [
        {
            'name': perusahaan['name'],
            'domain': perusahaan['domain'],
            'country': perusahaan['country']
        }
        for perusahaan in sorted(company_detail_list, key=lambda x: x['name'], reverse=True)
    ]
    return reverse_sorted

print("Hasil Soal 1:")
print(json.dumps(soal_1(), indent=4))
````

````python
from pprint import pprint
def soal_3():
    domain_karyawan = {}
    for perusahaan in company_detail_list:
        domain = perusahaan['domain']
        karyawan_di_domain = [
            emp['full_name'] for emp in employee_detail_list if emp['company'] == perusahaan['name']
        ]
        domain_karyawan[domain] = karyawan_di_domain
    return domain_karyawan

print("Hasil Soal 3:")
pprint(soal_3())
````

````python
import json

def soal_4():
    karyawan_negara_perusahaan = []
    for emp in employee_detail_list:
        # Cari negara dari perusahaan tempat karyawan bekerja
        negara_perusahaan = next((perusahaan['country'] for perusahaan in company_detail_list if perusahaan['name'] == emp['company']), None)
        karyawan_negara_perusahaan.append({
            'full_name': emp['full_name'],
            'company': emp['company'],
            'country': negara_perusahaan
        })
    return karyawan_negara_perusahaan

print("Hasil Soal 4:")
print(json.dumps(soal_4(), indent=4))
````

````python
import json
from collections import defaultdict

def soal_5():
    hasil = []
    
    perusahaan_kebangsaan = defaultdict(lambda: defaultdict(int))

    for emp in employee_detail_list:
        perusahaan_kebangsaan[emp['company']][emp['nationality']] += 1
    
    for perusahaan in company_detail_list:
        hasil.append({
            "company": perusahaan['name'],
            "employee_nationality": dict(perusahaan_kebangsaan[perusahaan['name']])
        })
    
    return hasil

print("Hasil Soal 5:")
print(json.dumps(soal_5(), indent=4))
````
